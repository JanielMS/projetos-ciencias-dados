```python
import pandas as pd 
import altair as alt 
```

```python
#alt.renderers.enable('mimetype')
```

```python
df = pd.read_csv('run_1km.csv', parse_dates=['startTime'], index_col=[0])
df['day_of_week'] = df['startTime'].dt.day_name()
df.head()
```

```python
alt.Chart(df).mark_point().encode(
    x='seconds_per_km:Q',
    y='day_of_week:N',
    color='day_of_week:N').interactive()
```

```python
# T: temporal type; N: nominal type; Q: quantitative type; O: ordinal type
alt.Chart(df).mark_point().encode(
    alt.X('seconds_per_km:Q'),
    alt.Y('day_of_week:N')
).interactive()
```

```python
# average, mean, median, stdev, min, max, sum
alt.Chart(df).mark_point().encode(
    x='mean(seconds_per_km)',
    y='day_of_week'
)
```

```python
alt.Chart(df).mark_point().encode(
    x='average(seconds_per_km)',
    y='day_of_week'
)
```

```python
alt.Chart(df).mark_point().encode(
    x='median(seconds_per_km)',
    y='day_of_week'
)
```

```python
alt.Chart(df).mark_point(color='#BBBBBB').encode(
    x='seconds_per_km',
    y='day_of_week'
)
```

```python
alt.Chart(df).mark_point(color='blueviolet').encode(
    x='seconds_per_km',
    y='day_of_week'
)
```

```python
alt.Chart(df).mark_point(color='#5f9ea0').encode(
    x='seconds_per_km',
    y='day_of_week'
)
```

```python
alt.Chart(df).mark_point(color='#5f9ea0').encode(
    alt.X('average(seconds_per_km)', title='Seconds/km (avg)', scale=alt.Scale(type='log')),
    alt.Y('day_of_week', axis=alt.Axis(title='Day of Week'))
)
```

```python
df['weekday'] = df['startTime'].dt.weekday
df.head(3)
```

```python
line = alt.Chart(df).mark_line(color='lightpink').encode(
    alt.X('weekday', title='Week Day Number'),
    alt.Y('average(seconds_per_km)'))

circle = alt.Chart(df).mark_circle(color='deeppink').encode(
    alt.X('weekday', title='Week Day Number'),
    alt.Y('average(seconds_per_km)'))

line + circle
```

```python
base = alt.Chart(df).mark_line(color='lightpink').encode(
    alt.X('weekday', title='Week Day Number'),
    alt.Y('average(seconds_per_km)'))

base + base.mark_circle(color='magenta')
```

```python
alt.Chart(df).mark_line(color='lightpink', point=True).encode(
    alt.X('weekday', title='Week Day Number'),
    alt.Y('average(seconds_per_km)'))
```

```python
speed = alt.Chart(df).mark_line(color='#A9A9A9', point=True).encode(
    alt.X('weekday', title='Week Day Number'),
    alt.Y('average(seconds_per_km)')
)

#dist = alt.Chart(df).mark_line().encode(
#    alt.X('weekday'),
#    alt.Y('sum(distance(m))')
#)

dist = speed.encode(
    alt.Y('sum(distance(m))')
)

dist | speed
```
